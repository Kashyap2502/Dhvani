```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def model(t, state, a, b, c):
    x, y, z = state
    dx=a*(y-b)
    dy=b*x-y-x*z
    dz=x*y-c*z

     
    return [dx, dy, dz]
 
a = 10.0
b = 28.0
c = 2.667
 
p = (a,b,c)  
 
y0 = [0.0, 1.0, 1.05] 
```

```python
t = np.arange(0.0, 40.0, 0.01)
 
result_odeint = odeint(model, y0, t, p, tfirst=True)

 
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot(result_odeint[:, 0],
        result_odeint[:, 1],
        result_odeint[:, 2],)
plt.show()
 
```
